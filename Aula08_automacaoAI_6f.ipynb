### Organizador de Arquivos por Extensão

Este script Python irá organizar os arquivos em um diretório especificado, movendo-os para subpastas que levam o nome de suas extensões (por exemplo, `.pdf`, `.jpg`). Ele utiliza as bibliotecas `os` para manipulação de caminhos e `shutil` para mover os arquivos.

**Instruções de Uso:**
1.  **Defina `source_directory`**: Altere a variável `source_directory` para o caminho da pasta que você deseja organizar.
2.  **Execute a célula**: O script criará subpastas (se não existirem) e moverá os arquivos para elas.

```python
import os
import shutil

# --- Configuração ---
# Defina o diretório que você deseja organizar
# EX: r'C:\Users\SeuUsuario\Documentos\Bagunca' ou '/home/seuusuario/Downloads'
source_directory = '/content/sample_data' # Mude este caminho para o seu diretório de bagunça

# --- Início da Execução ---
# Verifica se o diretório de origem existe antes de tentar organizá-lo
if not os.path.exists(source_directory):
    print(f"Erro: O diretório '{source_directory}' não existe. Verifique o caminho e tente novamente.")
else:
    print(f"Iniciando a organização de arquivos em: {source_directory}")

    # Itera sobre todos os itens (arquivos e pastas) no diretório de origem
    for filename in os.listdir(source_directory):
        # Cria o caminho completo para o item atual (arquivo ou pasta)
        file_path = os.path.join(source_directory, filename)

        # Verifica se o item atual é um arquivo (ignora subpastas para não movê-las)
        if os.path.isfile(file_path):
            # Extrai o nome base do arquivo e sua extensão
            base_name, extension = os.path.splitext(filename)

            # Determina o nome da pasta de destino com base na extensão
            if not extension: # Se o arquivo não tiver extensão (ex: 'Makefile')
                destination_folder_name = 'no_extension' # Move para uma pasta 'no_extension'
            else:
                # Remove o ponto da extensão para usar como nome da pasta (ex: '.pdf' -> 'pdf')
                destination_folder_name = extension[1:]

            # Cria o caminho completo para a pasta de destino (ex: '/content/sample_data/pdf')
            destination_folder_path = os.path.join(source_directory, destination_folder_name)

            # Se a pasta de destino ainda não existir, ela é criada
            if not os.path.exists(destination_folder_path):
                os.makedirs(destination_folder_path)
                print(f"Pasta '{destination_folder_name}' criada para receber os arquivos.")

            # Cria o caminho completo para o arquivo no seu novo destino
            destination_file_path = os.path.join(destination_folder_path, filename)

            try:
                # Move o arquivo do caminho de origem para o caminho de destino
                shutil.move(file_path, destination_file_path)
                print(f"Movido: '{filename}' para a pasta '{destination_folder_name}'")
            except shutil.Error as e:
                # Em caso de erro ao mover (ex: arquivo já existe no destino)
                print(f"Erro ao mover '{filename}': {e}")
            except Exception as e:
                # Captura outras exceções inesperadas durante o processo de movimentação
                print(f"Ocorreu um erro inesperado ao mover '{filename}': {e}")

    print("Organização concluída! Todos os arquivos foram processados.")
```

Este script Python irá organizar os arquivos em um diretório especificado, movendo-os para subpastas que levam o nome de suas extensões (por exemplo, `.pdf`, `.jpg`). Ele utiliza as bibliotecas `os` para manipulação de caminhos e `shutil` para mover os arquivos.

**Instruções de Uso:**
1.  **Defina `source_directory`**: Altere a variável `source_directory` para o caminho da pasta que você deseja organizar.

```python
import os

source_directory = '/content/sample_data'

# Lista de extensões e o número de arquivos para cada uma (10 de cada tipo)
extensions = {
    'txt': 10,
    'csv': 10,
    'jpg': 10,
    'pdf': 10
}

print(f"Gerando arquivos de teste em: {source_directory}")

# Cria os arquivos de teste
for ext, count in extensions.items():
    for i in range(1, count + 1):
        filename = f"file_{ext}_{i}.{ext}"
        file_path = os.path.join(source_directory, filename)
        # Cria um arquivo vazio
        with open(file_path, 'w') as f:
            f.write(f"Este é o arquivo {i}.{ext}\n")
        print(f"Criado: {filename}")

print("Geração de arquivos de teste concluída!")
```

### Consulta de Endereço por CEP (ViaCEP)

Este script Python permite consultar informações de endereço a partir de um CEP (Código de Endereçamento Postal) utilizando a API pública do ViaCEP. Ele retorna os dados formatados em JSON e inclui tratamento de erros para situações como CEP inválido, problemas de conexão ou falhas na API.

**Instruções de Uso:**
1.  Execute a célula de código abaixo.
2.  Utilize a função `consultar_cep` passando o CEP desejado como argumento.
3.  O resultado será um dicionário Python (que pode ser facilmente convertido para JSON) contendo os dados do endereço ou uma mensagem de erro.

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço.

    Args:
        cep (str): O CEP a ser consultado (apenas números).

    Returns:
        dict: Um dicionário contendo os dados do endereço ou uma mensagem de erro.
    """
    # Remove quaisquer caracteres não numéricos do CEP
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP possui o formato correto (8 dígitos)
    if len(cep) != 8:
        return {"erro": "CEP inválido. Deve conter 8 dígitos numéricos."}

    # URL da API ViaCEP
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Faz a requisição HTTP GET para a API
        response = requests.get(url)
        response.raise_for_status()  # Lança uma exceção para códigos de status HTTP ruins (4xx ou 5xx)

        # Decodifica a resposta JSON
        data = response.json()

        # Verifica se a API retornou um erro (por exemplo, CEP não encontrado)
        if 'erro' in data and data['erro']:
            return {"erro": f"CEP '{cep}' não encontrado ou inválido."}
        else:
            return data

    except requests.exceptions.HTTPError as http_err:
        # Erro de HTTP (ex: 404 Not Found, 500 Internal Server Error)
        return {"erro": f"Erro HTTP ao consultar o CEP '{cep}': {http_err}"}
    except requests.exceptions.ConnectionError as conn_err:
        # Erro de conexão (ex: sem internet)
        return {"erro": f"Erro de conexão ao consultar o CEP '{cep}': {conn_err}"}
    except requests.exceptions.Timeout as timeout_err:
        # Erro de tempo limite da requisição
        return {"erro": f"Tempo limite excedido ao consultar o CEP '{cep}': {timeout_err}"}
    except requests.exceptions.RequestException as req_err:
        # Outros erros de requisição
        return {"erro": f"Ocorreu um erro inesperado na requisição para o CEP '{cep}': {req_err}"}
    except json.JSONDecodeError:
        # Erro ao decodificar a resposta JSON (se a resposta não for um JSON válido)
        return {"erro": f"Erro ao decodificar a resposta JSON para o CEP '{cep}'."}
    except Exception as e:
        # Captura qualquer outra exceção inesperada
        return {"erro": f"Ocorreu um erro desconhecido ao consultar o CEP '{cep}': {e}"}
```

```python
# --- Demonstração de Uso Interativa ---

while True:
    user_cep_input = input("Digite o CEP para consulta (apenas números, ou 'sair' para encerrar): ")

    if user_cep_input.lower() == 'sair':
        print("Consulta encerrada.")
        break

    # Chama a função consultar_cep com o input do usuário
    resultado_consulta = consultar_cep(user_cep_input)

    # Imprime o resultado formatado em JSON
    print(json.dumps(resultado_consulta, indent=4, ensure_ascii=False))

    print("\n" + "-"*30 + "\n")
```
